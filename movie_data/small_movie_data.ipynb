```python
import pandas as pd
```

```python
movies_metadata = pd.read_csv('movies_metadata.csv')
credits = pd.read_csv('credits.csv')
keywords = pd.read_csv('keywords.csv')
links_small = pd.read_csv('links_small.csv')
links_small = links_small.rename(columns={"imdbId": "imdb_id"})
```

```python
links_small = links_small["imdb_id"].map(lambda x: "tt0" + str(x))
```

```python
credit_key = pd.merge(credits, keywords, how='outer')
credit_key
```

```python
new_movies = movies_metadata.drop([19730, 29503, 35587])
new_movies["id"] = pd.to_numeric(new_movies["id"])
```

```python
all_df = pd.merge(new_movies, credit_key)
all_df
```

```python
all_df = all_df[all_df["status"] == "Released"]
```

```python
small_df = pd.merge(all_df, links_small)
```

```python
small_df
```

```python
c = "belongs_to_collection, genres, imdb_id, original_language, overview, popularity, release_date, runtime, tagline, title, vote_average, cast, crew, keywords" 
filtered_columns = c.split(", ")
small_df = small_df[filtered_columns]
```

```python
import numpy as np
small_df["belongs_to_collection"] = small_df["belongs_to_collection"].fillna("{}")
small_df["genres"] = small_df["genres"].fillna("{}")
small_df["cast"] = small_df["cast"].fillna("{}")
small_df["crew"] = small_df["crew"].fillna("{}")
small_df["keywords"] = small_df["keywords"].fillna("{}")
small_df
```

```python
import ast
test = small_df
convert_columns = ["belongs_to_collection", "genres", "cast", "crew", "keywords"]

for convert in convert_columns:
    test[convert] = test[convert].map(lambda x: ast.literal_eval(x))
test
```

```python
test.to_json('small_movie_data.json', orient="index")
```
